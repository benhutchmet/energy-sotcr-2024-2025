```python
import requests
import pandas as pd
```

```python
# Load the fault data
faults = pd.read_csv("/home/users/benhutch/energy-sotcr-2024-2025/faults/fault_data/SPEN/all-lv-faults-spd-spm.csv")

# Read the head of the data
print(faults.head())
```

```python
# Subset to reporting year 2024
faults_2024 = faults[faults['Reporting Year'] == 2024]
```

```python
faults_2024.tail()
```

```python
# order from the higheest value of "Customers Interrupted" to the lowest
faults_2024 = faults_2024.sort_values(by='Customers Interrupted', ascending=False)
```

```python
faults_2024.head()
```

```python
# print all of the unique main causes
main_causes = faults_2024['Main Cause'].unique()

print(main_causes)

# Set up the met causes
met_causes = [
    "Wind and Gale (excluding Windborne Material)",
    "Rain",
    "Snow, Sleet and Blizzard",
    "Ice",
    "Falling live trees (not felled)",
    "Growing or Falling Trees (not felled)",
    "Lightning",
    "Falling dead trees (not felled)",
    "Freezing Fog and Frost",
    "Corrosion due to atmosphere/environment",
    "Flooding",
]
```

```python
# subset the faults 2024 to met causes
met_faults_2024 = faults_2024[faults_2024['Main Cause'].isin(met_causes)]
```

```python
len(met_faults_2024)
```

```python
met_faults_2024.head()
```

```python
# Make sure "Date/Time of Incidents" is a datetime "2024-12-07T08:57:59+00:00"
met_faults_2024['Date/Time of Incidents'] = pd.to_datetime(met_faults_2024['Date/Time of Incidents'])
```

```python
# Plot the number of customers interrupted by time
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 5))

# Plot the number of customers interrupted by time
plt.plot(met_faults_2024['Date/Time of Incidents'], met_faults_2024['Customers Interrupted'], 'o')
```

```python
# Set up a dictionary for the storms
storms = {
    "Storm Ashley": ["2024-10-18", "2024-10-21"],
    "Storm Bert": ["2024-11-21", "2024-11-25"],
    "Storm Conall": ["2024-11-26", "2024-11-27"],
    "Storm Darragh": ["2024-12-05", "2024-12-07"],
    "Storm Éowyn": ["2025-01-21", "2025-01-24"],
}

# Set up different colours for the storms
storm_colours = {
    "Storm Ashley": "red",
    "Storm Bert": "blue",
    "Storm Conall": "green",
    "Storm Darragh": "purple",
    "Storm Éowyn": "orange",
}

# Redo the plot, but highlight the time periods of the storms
plt.figure(figsize=(10, 5))

# Plot the number of customers interrupted by time
plt.plot(met_faults_2024['Date/Time of Incidents'], met_faults_2024['Customers Interrupted'], 'o')

# Loop through the storms and plot the time periods
for storm, dates in storms.items():
    start_date = pd.to_datetime(dates[0])
    end_date = pd.to_datetime(dates[1])
    plt.axvspan(start_date, end_date, color=storm_colours[storm], alpha=0.5, label=storm)

# Set the title and labels
plt.title("Customers Interrupted by Storms in 2024, SPEN")

plt.xlabel("Date")
plt.ylabel("Customers Interrupted")

# Set the legend
plt.legend(loc='upper left')
```
