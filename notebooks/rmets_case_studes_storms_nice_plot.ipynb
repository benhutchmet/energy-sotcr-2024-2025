```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import iris
import cartopy.crs as ccrs
import cartopy.feature as cf
import matplotlib as mpl
from iris.coord_categorisation import add_day_of_year
from iris.coord_categorisation import _pt_date
from iris.time import PartialDateTime
import iris.quickplot as qplt
```

```python
start_years = [2023,2023,2023,2023,2023,2023,2023,2023,2024,2024,2024,2024]
start_months = [4,9,10,11,11,12,12,12,1,1,1,4]
start_days = [11,26,18,1,12,9,20,27,2,20,21,6]

end_years = [2023,2023,2023,2023,2023,2023,2023,2023,2024,2024,2024,2024]
end_months = [4,9,10,11,11,12,12,12,1,1,1,4]
end_days = [13,28,20,3,14,11,22,29,4,22,23,8]

storm_name = ['Noa: 12/04/23','Agnes: 27/09/23','Babet: 19/10/23','Ciaran: 02/11/23', 'Debi: 13/11/23','Elin/Fergus: 10/12/23','Pia: 21/12/23','Gerrit: 28/12/23','Henk: 3/1/24','Isha: 21/1/24','Jocelyn: 22/1/24','Kathleen: 7/4/24' ]



data_dir = '/gws/pw/j07/ceraf/hbloomfield01/Data/HYDRA/ERA5/'
data_str = 'ERA5_EU_extr_weather_' # YYYY_MM.nc

# constraints
u10= iris.Constraint('i10fg')
levels=[0,20,22.5,25,27.5,30,32.5,35,37.5,40]
cmap=plt.cm.get_cmap('hot_r')




fig = plt.figure(figsize=(12,8))


for i in range(0,len(start_years)):
    dt_start = datetime.datetime(start_years[i],start_months[i],start_days[i]) 
    dt_end = datetime.datetime(end_years[i],end_months[i],end_days[i]) 

    event_constraint = iris.Constraint(time=lambda cell: dt_start <= cell.point <= dt_end)

    # load in whole dataset over the event.
    dataset_event = iris.load(data_dir + data_str  + str(start_years) + '*.nc',event_constraint)
    event_u10 = dataset_event.extract(u10)
    event_conc = event_u10.concatenate_cube()
    event_footprint =  event_conc.collapsed('time',iris.analysis.MAX) 
    LONS,LATS = iris.analysis.cartography.get_xy_grids(event_footprint)

    ax = plt.subplot(3,4,i+1,projection=ccrs.PlateCarree())
    ax.coastlines(resolution='50m')
    ax.set_extent([-11,2,49,60]) # W E S N 
    x = plt.pcolormesh(LONS,LATS,event_footprint.data,cmap=cmap,transform=ccrs.PlateCarree(),norm = mpl.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=False))  
    plt.title(storm_name[i],fontsize=14)
    ax.add_feature(cf.BORDERS)

    #c = plt.colorbar(x, orientation='horizontal',ticks=[0,20,25,30,35,40,45,50],extend='max')
    #c.ax.tick_params(labelsize=14)
    #c.set_label('Max. 72 hour 10m wind gust',fontsize=12)

fig.subplots_adjust(bottom=0.25)
cbar_ax = fig.add_axes([0.17, 0.17, 0.65, 0.04])
c = fig.colorbar(x, cax=cbar_ax, orientation="horizontal",ticks=[0,20,25,30,35,40,45,50],extend='max')
c.ax.tick_params(labelsize=14)
c.set_label('Max. 72 hour 10 m wind gust (metres per second)',fontsize=14)

plt.savefig('/home/users/hbloomfield01/rmets_report/storm_footprints_23_24.png',dpi=600)
plt.close()
```
